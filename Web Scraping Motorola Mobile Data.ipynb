# Importing Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
link="https://www.flipkart.com/search?q=moto+g31&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_3_5"
data=requests.get(link)
# soup=BeautifulSoup(data,"html")
```

```python
soup1=BeautifulSoup(data.content,"html.parser")

soup1.prettify
```

```python
elements=soup1.find_all('div',class_="_4rR01T")
lis1=[]
for element in elements:
    lis1.append(element.get_text())
#     print(element.get_text())
lis1
```

```python
prices=soup1.find_all('div',class_="_30jeq3 _1_WHN1")
lis2=[]
for price in prices:
    lis2.append(price.get_text()[1:])
lis2
```

```python
ratings=soup1.find_all('div',class_="_3LWZlK")
lis3=[]
for rating in ratings[0:24]:
    lis3.append(rating.get_text())
lis3
```

```python
len(lis1),len(lis2),len(lis3)
```

```python
df=pd.DataFrame({'Model':lis1,'Price in INR':lis2,'Ratings':lis3})
df
```
